```python
import pandas as pd
from splinter import Browser
from bs4 import BeautifulSoup
```

```python
# Set up Splinter
browser = Browser('chrome')
```

```python
# Visit the Wine Folly "40 Wine Descriptions" article
url = "https://winefolly.com/tips/40-wine-descriptions/"
browser.visit(url)
```

```python
# Parse the website
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
```

```python
# Create a variable for the main area of the list
article_dts=soup.find("dl").find_all("dt")


print(article_dts)
```

```python
# Convert the list into a dataframe
description_df = pd.DataFrame(article_dts)
description_df.head()
```

```python
# Convert the descriptions to lowercase
description_df[0]= description_df[0].apply(lambda x: x.lower())
description_df
```

```python
# Export dataframe to json file
description_df.to_json("Resources/wine_descriptions.json")
```
